## Handling Missing Value

```python
df['Original_price'].describe()
```

```python
# Cek Outliers di Original_price

IQR = 980 - 273
lower_fence = 273 - (1.5 * IQR)
upper_fence = 980 + (1.5 * IQR)
df[(df['Original_price'] < lower_fence) | (df['Original_price'] > upper_fence)]
```

- Karena di kolom Original_price terdapat beberapa outliers maka missing value akan di fill dengan median

```python
df['Original_price'].fillna(df['Original_price'].median(), inplace=True)
```

```python
df['perc_of_unit_sold'].describe()
```

```python
## Cek outliers di perc_of_unit_sold

IQR = 79 - 36.25
lower_fence = 36.25 - (1.5 * IQR)
upper_fence = 79 + (1.5 * IQR)
df[(df['perc_of_unit_sold'] < lower_fence) | (df['perc_of_unit_sold'] > upper_fence)]
```

- di kolom perc_of_unit_sold tidak ada outliers. Maka missing value akan diisi dengan mean

```python
df['perc_of_unit_sold'].fillna(df['perc_of_unit_sold'].mean(), inplace=True)
```

```python
df['rating'].describe()
```

```python
## Cek outliers di rating

IQR = 96 - 82
lower_fence = 82 - (1.5 * IQR)
upper_fence = 96 + (1.5 * IQR)
df[(df['rating'] < lower_fence) | (df['rating'] > upper_fence)]
```

- Karena di kolom rating terdapat beberapa outliers maka missing value akan di fill dengan median

```python
df['rating'].fillna(df['rating'].median(), inplace=True)
```

```python
df.shpping_rate.describe()
```

```python
df['shpping_rate'].fillna('FREE Shipping', inplace=True)
```

```python
df.gender.describe()
```

```python
df['gender'].fillna('men', inplace=True)
```

```python
df.isna().sum()
```

- Missing value di kolom shpping_rate dan gender diisi dengan most frequent, karena jumlah missing value nya sedikit, sayang jika di drop semua

## Kategori barang yang laku dijual di e-commerce, Berdasarkan avg persentase barang terjual

```python
pd.crosstab(index=df['category'], columns='avg % barang terjual', values=df['perc_of_unit_sold'], aggfunc='mean').sort_values('avg % barang terjual', ascending=False)
```

- Dilihat dari avg persentase barang terjual, kategori Perfumes & Fragrances punya rata2 persentase barang terjual yang tinggi.

```python
parfum = df[df['category'] == 'Perfumes & Fragrances']
```

- Kita akan coba perdalam dengan cek 10 parfum yang % barang terjual nya tinggi

```python
pd.crosstab(index=[parfum['category'], parfum['product'], parfum['gender']], columns='% barang terjual', values=parfum['perc_of_unit_sold'], aggfunc='mean').sort_values('% barang terjual', ascending=False).head(10)
```

- Ternyata parfum untuk pria mendominasi 10 parfum dengan % barang terjual yang tinggi

## Mengetahui harga yang tepat untuk menjual barang

```python
pd.crosstab(index=df['category'], columns='avg harga', values=df['Original_price'], aggfunc='mean').round().sort_values('avg harga', ascending=False)
```

```python
pd.crosstab(index=df['category'], columns='avg harga', values=df['price'], aggfunc='mean').round().sort_values('avg harga', ascending=False)
```

- Terdapat perbedaan pada rata2 harga Perfumes dan Sportswear antara harga sebelum diskon dan sesudah diskon
- Sebelum diskon harga rata2 parfum lebih tinggi dibandingkan Sportswear, tetapi setelah diskon harga rata2 Perfumes menjadi lebih rendah. Berarti Perfumes diberi diskon yang cukup besar dibanding Sportswear

- Lebih detail lagi kita coba cek perbandingan rata2 harga parfum untuk women, men, dan unisex

```python
pd.crosstab(index=[parfum['category'], parfum['gender']], columns='avg harga', values=parfum['price'], aggfunc='mean').round().sort_values('avg harga', ascending=False).head(10)
```

- Ternyata rata2 harga parfum wanita lebih mahal dibandingkan parfum pria dan unisex
